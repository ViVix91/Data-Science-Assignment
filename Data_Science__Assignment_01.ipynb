Checking the number of matches played and won by each team (sorted according to win percentage)

Function to get the scorecard of each match

```python
def get_scorecard(innings):

    scorecards = []

    for inn in innings:
        inn_data = list(inn.values())[0]
        team = inn_data['team']
        deliveries = inn_data['deliveries']

        batsman_runs = {}
        bowler_stats = {}

        for ball in deliveries:
            delivery = list(ball.values())[0]
            batsman = delivery['batsman']
            bowler = delivery['bowler']
            runs = delivery['runs']['batsman']
            total_runs = delivery['runs']['total']

            batsman_runs[batsman] = batsman_runs.get(batsman, 0) + runs

            if bowler not in bowler_stats:
                bowler_stats[bowler] = {'runs':0,'wickets':0}
            bowler_stats[bowler]['runs'] += total_runs
            if 'wicket' in delivery:
                bowler_stats[bowler]['wickets'] += 1

        batsman_df = (
            pd.DataFrame(batsman_runs.items(), columns=['Batsman','Runs'])
            .sort_values('Runs', ascending=False)
            .head(4)
            .reset_index(drop=True)
        )
        bowler_df = (
            pd.DataFrame(
                [(k,v['runs'],v['wickets']) for k,v in bowler_stats.items()],
                columns=['Bowler','Runs Conceded','Wickets']
            )
            .sort_values(['Wickets','Runs Conceded'], ascending=[False,True])
            .head(4)
            .reset_index(drop=True)
        )
        scorecard = pd.concat([batsman_df, bowler_df], axis=1)
        scorecards.append((team, scorecard))

    return scorecards
```

```python
all_scorecards = []

for i in range(len(df)):
    innings_data = ast.literal_eval(df['innings_data'].iloc[i])
    match_scorecards = get_scorecard(innings_data)
    all_scorecards.append(match_scorecards)
```
